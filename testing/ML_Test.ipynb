```python
import os
import sys
notebook_dir = os.getcwd()
sys.path.append(os.path.abspath(os.path.join(notebook_dir, '..')))
from testing.LSTM_Test import TestingLSTM_Model
from testing.GRU_Test import testingGRU_Model
from testing.RF_Test import testingRF_Model
from models.FindTime import preprocessing
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
location = "WARRIGAL_RD N of HIGH STREET_RD"
csv_path = "../csv/ScatsReformed.csv"
TOD = "Morning"

X, y, scaler, TOD = preprocessing(csv_path, location, lookback=10, TOD=TOD)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)
```

<h1>Base Configs</h1>

```python
print("LSTM")
lstm_pred, lstm_true, TOD = TestingLSTM_Model(X_train, X_test, y_train, y_test, scaler, 10)


plt.plot(lstm_true, label="Actual")
plt.plot(lstm_pred, label="LSTM")
plt.title(f"Predictions for {location} ({TOD})")
plt.legend()
plt.show()
```

```python
print("GRU")
gru_pred, gru_true, TOD = testingGRU_Model(X_train, X_test, y_train, y_test, scaler, 10)
plt.plot(gru_true, label="Actual")
plt.plot(gru_pred, label="GRU")
plt.title(f"Predictions for {location} ({TOD})")
plt.legend()
plt.show()
```

```python
print("RF")
rf_pred, rf_true, TOD = testingRF_Model(X_train, X_test, y_train, y_test, scaler, 100)
plt.plot(rf_true, label="Actual")
plt.plot(rf_pred, label="RF")
plt.title(f"Predictions for {location} ({TOD})")
plt.legend()
plt.show()
```
